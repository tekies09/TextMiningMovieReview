```python
def get_data_frame(url) :
    import bs4
    import requests as rq
    import pandas as pd
    from pandas import DataFrame as df
    list_review=[]
    list_score =[]
    html= rq.get(url).text
    soup = bs4.BeautifulSoup(html,'lxml')
    score_result = soup.find('div',class_='score_result')
    lis = score_result.findAll('li')
    for li in lis :
        review_text = li.find('div',class_='score_reple').find_all('p')[0].getText().replace('관람객','')
        review_score = li.find('em').getText()
        list_review.append(review_text)
        list_score.append(review_score)
    df1 =df(data={'text':list_review,'score':list_score},columns=['text','score'])
    return df1
```

```python
def get_data(url) :
    import bs4
    import requests as rq
    list_review=[]
    html= rq.get(url).text
    soup = bs4.BeautifulSoup(html,'lxml')
    score_result = soup.find('div',class_='score_result')
    lis = score_result.findAll('li')
    for li in lis :
        review_text = li.find('div',class_='score_reple').find_all('p')[0].getText().replace('관람객','')
        list_split = review_text.split()
        list_review = list_review + list_split
    return list_review
# 사용을 위해 함수로 만듬 
```

```python
import pandas as pd
from pandas import DataFrame as df
import bs4
import requests as rq
test_url ="https://movie.naver.com/movie/bi/mi/pointWriteFormList.nhn?code=182205&type=after"
html= rq.get(test_url).text
soup = bs4.BeautifulSoup(html,'lxml')
a=soup.find_all('strong',class_='total')[0].find_all('em')[0].getText()
total_count = int(a.replace(',', ''))
print(total_count)
list_review=[]
for i in range(1, int(total_count / 10) + 1):
    url = test_url + '&page=' + str(i)
    print('url: "' + url + '" is parsing....')
    list_review=list_review+get_data(url)
dataframe =pd.DataFrame(list_review)
dataframe.to_csv("review.csv",encoding='euc-kr',header=False,index=False)
#각 페이즈를 세서 페이지마다 리뷰를 저장함 
```
